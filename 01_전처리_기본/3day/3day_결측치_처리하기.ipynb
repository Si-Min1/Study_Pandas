### 결측치 대체하기

```python
# 결측치를 모두 0 으로 채우기
df.fillna(0)
# df.fillna( ' ')
```

```python
# 이용거리이라는 컬럼안에 결측치가 있을경우 0으로 바꿔라
df['이용거리'].fillna(0)

# df['이용거리'].fillna(0)[df['이용거리'] == 0] # 확인용
```

```python
# 이용거리중 결측치를 'missing'이라고 수정하시오
df['이용거리'].fillna('missing')
# float타입이 object타입으로 바뀜
```

```python
# 평균으로 대체값 사용하기
# 전체 데이터 중에 결측치가 있으며
# 전체 평균값으로 채워주세오

df.fillna(df.mean())
# 경고 메시지 뜸(님 문자열 있음 일단 뺌, 똑디 고르셈)
```

```python
# 이용거리 결측치 이용거리 평균으로 채우기

df['이용거리'].fillna(df['이용거리'].mean())
```
