### tips 데이터셋 읽어들이기

```python
import pandas as pd

# seaborn과 연동되어 있는 모듈들
# 그림 그리는 라이브러리
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns

#행의 위치 찾을 때 사용
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips
```

### 빈도분석

```python
tips[['sex']]
```

```python
# 성별 빈도 분석

tips['sex'].value_counts()
# tips['sex'].count() # 이건 데이터 갯수만 확인
```

```python
# 요일 및 흡연 여부 시간 빈도 세기
print(tips['smoker'].value_counts(),'\n')
print(tips['day'].value_counts(),'\n')
print(tips['time'].value_counts())
```

```python
# 월별.성별 교차분석
pd.crosstab(tips['sex'], tips['day'])
```

```python
# 월별.성별 교차분석 총괄표
pd.crosstab(tips['sex'], tips['day'], margins = True)
```

```python
# 빈도 비율 확인
# r에 데이터의 값이 들어간다 (ex. 목요일 + 남자 = 30)
# 그 후 30(위의 경우) / len(tips) = 전체 데이터에 대한 비율
pd.crosstab(tips['sex'],
            tips['day']).apply(
                lambda r:r/len(tips) * 100,
                axis = 1
            )
```
