# 2. 재 구조화

```python
# cut() 동일한 길이(값의 범위)로 3구간으로 나눈다
df1 = pd.cut(df['math'],3)
df1
# 맨 아래 카테고리에 적힌 3구간으로 나누고 나뉜 구간을 위에 데이터 별로 옆에 적어줌
```

```python
df2 = df['math'].groupby(df1)
# 각 구간별 1명씩 보여줌 (1구간 2구간 3구간에 들어가는 사람 1명씩)
df2.head(1)
```

```python
# df2에 대한 기초통계
# 몇명, 평균, 
df3 = df2.agg(['count','mean','std','min','max'])
df3
```

```python
pd.qcut(df['math'],3)
```

```python
df4 = pd.qcut(df['math'],3, labels = False)
df4
```

```python
import numpy as np
```

```python
# 결과값 반대로 보기
df5 = pd.qcut(df['math'],3, labels = np.arange(3,0,-1))
df5
```

```python
df6 = df['math'].groupby(df5)
df6.head(1)
```

```python
# 기초통계 조회
# count, mean, std, min, max
# df6.describe()
# 2구간에 들어가는 학생 수가 1명이라 중앙값은 nan이 뜬다
df6.agg(['count','mean','std','min','max'])
```
