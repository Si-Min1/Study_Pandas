## 2. numpy의 loc과 iloc 사용하기

```python
# 차이점을 확인하기 위해 특정 행 하나를 삭제
# 안써도 컬럼 안짜를거면 안써도 될듯
tips_del = tips.drop([2])
tips_del
```

```python
# iloc는 행의 위치(순서)로 체크한다 그냥 배열위치
tips_del.iloc[2] # <= 는 tips_del 에서 인덱스가 3인 곳에 위치하는 데이터 이다. 데이터는 위에 참조
```

## 데이터 병합하기 (합치기)

```python
# 시리즈 데이터 합하기
# 시리즈변수명 s1 ==> 값은 0, 1 / 인덱스 값은 A, B로 시리즈 만들기
# 시리즈변수명 s2 ==> 값은 2, 3, 4 / 인덱스 값은 a, b, c로 시리즈 만들기


# sd1 = pd.Series(['Dog', 'Cat', 'Tiger', 'Lion', 'Monkey'],
#                index = [1,2,6,4,5]) # 그냥 짜도 됨

s1 = {'A' : 0, 'B' : 1}
s2 = {'a' : 2, 'b' : 3, 'c' : 4}

s1 = pd.Series(s1)
s2 = pd.Series(s2)
print(s1,'\n's2)
```

```python
# 행을 병합하기
pd.concat([s1,s2], axis = 0)
```

### 데이터프레임 JOIN

```python
df_info1 = pd.DataFrame({'고객번호' : [1001, 1002, 1003, 1004, 1005, 1006, 1007],
                         '이름' : ['강', '김', '홍', '이', '박', '최', '소']},
                       columns = ['고객번호','이름'])
df_info1
```

```python
df_info2 = pd.DataFrame({'고객번호' : [1001, 1001, 1005, 1006, 1008, 1001],
                         '금액' : [10000, 20000, 15000, 5000, 10000, 30000]},
                       columns = ['고객번호','금액'])
df_info2
```

```python
# 컬럼 단위로 추가
# 단, 고객 번호가 같은 값들만

print(pd.merge(df_info1, df_info2, how='inner'),"\n")

print(pd.merge(df_info2, df_info1, how='inner'),"\n")
```

```python
print(pd.merge(df_info1, df_info2, how='outer'), '\n\n') # 서로 들고 있는거 다 내놓음

print(pd.merge(df_info1, df_info2, how='left'), '\n\n') # 왼쪽은 1008를 모름

print(pd.merge(df_info1, df_info2, how='right')) # 오른쪽은 1001, 1005, 1006, 1008 말고는 모름
```
