```python
# ans 데이터 프레임을 사용합니다
# dataset 데이터 값 중에 I(로마자 1)인 값을 조회해 주세요
# 변수는 fillter_row_chk 사용
fillter_row_chk = (ans['dataset'] == 'I')
fillter_row_chk.head(2)
```

```python
# 필터링 된 데이터를 조회하기
# 변수는 data1
data1 = ans[fillter_row_chk]
data1
```

```python
# data1.plot()

# 선그래프 그리기
plt.plot(data1['x'], data1['y'])
```

```python
# 점 찍기
plt.plot(data1['x'], data1['y'], 'o')
```

```python
# x모양으로 점 찍기
plt.plot(data1['x'], data1['y'], 'x')
```

```python
# dataset 안에 있는 ll, lll, lv에 대해서도 필터링 후
# data2,3,4 변수에 각각 넣고 조회하시오
data2 = ans[(ans['dataset'] == 'II')]
print(data2)
data3 = ans[(ans['dataset'] == 'III')]
print(data3)
data4 = ans[(ans['dataset'] == 'IV')]
print(data4)
```

```python
# 시각화 옵션들
plt.title('Line')
plt.plot([1,2,3],[1,2,3],marker = 's')
plt.show()
```

```python
# 범례 추가하기
plt.title('legend')
plt.plot([1,2,3,4], label = 'asc')
plt.plot([4,3,2,1], label = 'desc')

# 범례 추가
plt.legend() # 여기서 라벨을 알려줌
plt.show()
```

```python
plt.plot([1,2,3],[1,2,3], color = 'red')
plt.show()
```

```python
plt.plot([1,2,3],[1,2,3], color = 'red')
# x,y 축 이름 만들어 주기
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.show()
```

```python
# 3개의 선 그리기
# 각 선의 색상, 선 스타일 변경
plt.plot([1,2,3,4], color='r', linestyle = '-', label='line')
plt.plot([4,3,2,1], color='g', linestyle = '--', label='dashed')
plt.plot([2,1,1,2], [3,4,4,3], color='b', linestyle = ':', label='dotted')

plt.legend()
plt.show()
```

```python
# x,y축의 범위 지정하기..
# 한글 깨짐 발생 폰트 지정
plt.rc('font', family = 'Malgun Gothic')

plt.title('x,y 레인지')
plt.plot([10, 20, 30, 40, 50],
         [1, 4, 9, 16, 24],
         c = 'b', lw = 5, ls = '--', marker = 'o', ms = 15,
         mec = 'g', mew = 5, mfc = 'r')
# c = 색상, lw = 줄 두께, ls = 라인 스타일, 마커는 마커스타일, ms는 마커 크기
# mec = 마커테두리 색깔, mew = 마커 테두리 사이즈, mfc 마커 색깔


# x축의 범위 지정
plt.xlim(0,55)

# y축의 범위 지정
plt.ylim(-10,30)

plt.show()
```

```python
# 붓꽃 데이터 셋
iris = sns.load_dataset('iris')
# 타이타닉호 데이터 셋
titanic = sns.load_dataset('titanic')
# 팁 데이터 셋
tips = sns.load_dataset('tips')
# 여객운송 데이터 셋
flights = sns.load_dataset('flights')
```

```python
print(iris, titanic, tips, flights)
```

```python
sns.set_palette('pastel') # gray , dark
sns.stripplot(data = tips, x='day', y= 'total_bill')
plt.title('팁_데이터셋')
plt.xlabel('요금')
plt.ylabel('요일')

plt.show
```

```python
# sns (seaborn)에서 제공하는 막대 그래프
# 중앙에 잇는 검은선 오차막대(error bar) : 오차막대의 길이는 작을수록 좋다
sns.barplot(data=tips, x = 'sex', y= 'total_bill')
plt.show()
```

```python
# 3개의 컬럼을 사용
# 기본x,y에 더해서
# 상세그룹을 더 추가하고자 할때 사용함

sns.barplot(data=tips, x = 'sex', y= 'tip', hue = 'day')
```

```python
# count 그래프 그리기 (빈도분석과 동일)
sns.catplot(data = tips, 'day', count='count')
```

```python
sns.boxplot(x='sex', y='tip', data = tips)
```

```python
# orient 'h는 수평 , v가 수직(디폴트값)'
sns.boxplot(data = tips, orient = 'h')
```

```python
# rug플롯 : 데이터 위치를 x축 위에 작은 선분(rug)로 나타내어
#           실제 데이터들의 위치를 보여줍니다.
sns.rugplot(tips['tip'], height = 0.5)
# 폰트오류는 위 맑은 고딕 폰트 쓰면 됨
# 어디에 몰렸는가 확인용이긴한데 딱히 안씀
```

```python
# 커널 밀도 그래프 통계하는 사람 가끔 쓴다 카더라
sns.kdeplot(tips['tip'])
sns.distplot(tips['tip'])
# 하나만 쓰면 멋이 안살아서 같이 쓰면 좋을듯
# 막대그래프를 벗어나는 값들은 이상치 데이터로 봐도 좋을듯
```
