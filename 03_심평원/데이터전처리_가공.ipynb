# 연령대 조건 만들기


```python
# 10세미만, 10대, 20대 ... 90대, 100세이상 으로 그룹 만들기
df_9under = df_month_age[(df_month_age['수진자연령'] < 10)]
df_9under = df_9under.reset_index(drop=True)
df_9under
# '연령대' 컬럼 추가 후 값음 모두 9세 이하 넣기
df_9under['연령대'] = '9세 이하'
df_9under
```

```python
# 위에 한거 df_10,df_20 형식으로 90대 까지 만드시오 (100세 이상은 없음)
df_10 = df_month_age[(df_month_age['수진자연령'] >= 10) & (df_month_age['수진자연령'] < 20)]
df_10 = df_10.reset_index(drop=True)
df_10['연령대'] = '10대'
df_20 = df_month_age[(df_month_age['수진자연령'] >= 20) & (df_month_age['수진자연령'] < 30)]
df_20 = df_20.reset_index(drop=True)
df_20['연령대'] = '20대'
df_30 = df_month_age[(df_month_age['수진자연령'] >= 30) & (df_month_age['수진자연령'] < 40)]
df_30 = df_30.reset_index(drop=True)
df_30['연령대'] = '30대'
df_40 = df_month_age[(df_month_age['수진자연령'] >= 40) & (df_month_age['수진자연령'] < 50)]
df_40 = df_40.reset_index(drop=True)
df_40['연령대'] = '40대'
df_50 = df_month_age[(df_month_age['수진자연령'] >= 50) & (df_month_age['수진자연령'] < 60)]
df_50 = df_50.reset_index(drop=True)
df_50['연령대'] = '50대'
df_60 = df_month_age[(df_month_age['수진자연령'] >= 60) & (df_month_age['수진자연령'] < 70)]
df_60 = df_60.reset_index(drop=True)
df_60['연령대'] = '60대'
df_70 = df_month_age[(df_month_age['수진자연령'] >= 70) & (df_month_age['수진자연령'] < 80)]
df_70 = df_70.reset_index(drop=True)
df_70['연령대'] = '70대'
df_80 = df_month_age[(df_month_age['수진자연령'] >= 80) & (df_month_age['수진자연령'] < 90)]
df_80 = df_80.reset_index(drop=True)
df_80['연령대'] = '80대'
df_90 = df_month_age[(df_month_age['수진자연령'] >= 90) & (df_month_age['수진자연령'] < 100)]
df_90 = df_90.reset_index(drop=True)
df_90['연령대'] = '90대'
```

```python
# df_month_age_any 데이터 프레임 새로 만들어서
# 위에 df_9under 부터 df_90까지 모두 행단위로 추가하시오
# 모두 추가 후 인덱스 번호 재배열 완료 후 조회

df_month_age_any = pd.DataFrame()
df_month_age_any = pd.concat([df_9under, df_10, df_20, df_30, df_40, df_50,
                              df_60, df_70, df_80, df_90]).reset_index(drop = True)
df_month_age_any
```

### 위 내용을 최대한 편하게 코드를 줄여보시오.

```python
df_month_age['연령대'] = '1'
for i in range((len(df_month_age))):
    if df_month_age['수진자연령'][i] < 10:
        df_month_age['연령대'][i] = '10세 미만'
    elif df_month_age['수진자연령'][i] >= 10 and df_month_age['수진자연령'][i] < 20:
        df_month_age['연령대'][i] = '10대'
    elif df_month_age['수진자연령'][i] >= 20 and df_month_age['수진자연령'][i] < 30:
        df_month_age['연령대'][i] = '20대'
    elif df_month_age['수진자연령'][i] >= 30 and df_month_age['수진자연령'][i] < 40:
        df_month_age['연령대'][i] = '30대'
    elif df_month_age['수진자연령'][i] >= 40 and df_month_age['수진자연령'][i] < 50:
        df_month_age['연령대'][i] = '40대'
    elif df_month_age['수진자연령'][i] >= 50 and df_month_age['수진자연령'][i] < 60:
        df_month_age['연령대'][i] = '50대'
    elif df_month_age['수진자연령'][i] >= 60 and df_month_age['수진자연령'][i] < 70:
        df_month_age['연령대'][i] = '60대'
    elif df_month_age['수진자연령'][i] >= 70 and df_month_age['수진자연령'][i] < 80:
        df_month_age['연령대'][i] = '70대'
    elif df_month_age['수진자연령'][i] >= 80 and df_month_age['수진자연령'][i] < 90:
        df_month_age['연령대'][i] = '80대'
    elif df_month_age['수진자연령'][i] >= 90:
        df_month_age['연령대'][i] = '90대'
df_month_age
# Pandas에서는 파생 DataFrame에 수정을 가하는 것을 권장하지 않습니다(그래서 경고를 띄우는 것이죠.) 
# 대신 Pandas는 copy() 메서드를 통해 파생 DataFrame에 독립적인 메모리를 부여한 뒤 여기에 수정을 가하도록 유도하지요

```
