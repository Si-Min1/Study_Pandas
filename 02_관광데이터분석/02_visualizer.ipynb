```python
import pandas as pd

# 시각화 라이브러리
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
# 시스템 OS 확인을 위한 라이브러리
import platform
```

```python
# 한글 깨짐 발생 폰트 지정
plt.rc('font', family = 'Malgun Gothic')
```

```python
# 운영체제별 폰트 정의하기
if platform.system() == 'Window':
    path = 'C:/Windows/Fonts/Malgun Gothic.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font',family = font_name)
elif platform.system() == 'Darwin':
    rc('font',family = 'AppleGothic')
    # plt.rcParams['axes.unicode_minus'] = False # Mac용
else:
    print("Check your OS system")
```

```python
# 중국 국적에 대한 데이터를 조회하기
df_filter = df[df['국적'] == '중국']
df_filter
```

```python
# 중국, 일본, 대만, 미국, 홍콩 각각에 대한
# 기준년월별 관광객 추이분석을 해보시오
country_list = ['중국', '일본', '미국', '대만', '홍콩']
for name in country_list:
    df_filter = df[df['국적'] == name]
    # 그래프 꾸미기
    # 그래프 크기 조절
    plt.figure(figsize = (12,4))

    # 그래프에 데이터 넣기 (위에거)
    plt.plot(df_filter['기준년월'], df_filter['관광'])

    # 그래프 제목
    plt.title(name + '국적 관광객 추이')

    # x, y축에 대한 라벨추가
    plt.xlabel('기준년월')
    plt.ylabel('관광객수')

    # X축에 대한 눈금을 조정
    plt.xticks(['2010-01', '2011-01', '2012-01', '2013-01', '2014-01',
                '2015-01', '2016-01', '2017-01', '2018-01', '2019-01'])

    plt.show()
```

```python
# 2.히트맵 그리기
# 히트맵은 시본 라이브러리를 함께 사용해야 합니다.
import seaborn as sns
```
